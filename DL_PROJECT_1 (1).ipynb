```python
import os
import pandas as pd

def list_image_paths(directory):
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tif', '.tiff']
    image_paths = []

    for root, _, files in os.walk(directory):
        for file in files:
            if any(file.lower().endswith(ext) for ext in image_extensions):
                image_paths.append(os.path.join(root, file))

    return image_paths

image_paths = list_image_paths(directory_path)

# Create a DataFrame
df = pd.DataFrame({'Image Paths': image_paths})

# Print the DataFrame
print(df)
```

```python
from PIL import Image, ImageOps, ImageEnhance, ImageFilter
import os


directory_path=("/content/drive/MyDrive/train dl")

preprocessed_images = []

# Loop through all files in the directory
for filename in os.listdir(directory_path):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        # Constructing the full path to the image file
        image_path = os.path.join(directory_path, filename)

        # Opening the image
        img = Image.open(image_path)

        # Applying preprocessing steps
        # 1. Resize the image
        img = img.resize((256, 256))

        # 2. Converting to grayscale
        img = ImageOps.grayscale(img)

        # 3. Normalizing pixel values to [0, 1]
        img = ImageOps.autocontrast(img)

        # 4. Applying Gaussian blur
        img = img.filter(ImageFilter.GaussianBlur(radius=1))

        # 5. Enhancing brightness and contrast
        enhancer = ImageEnhance.Brightness(img)
        img = enhancer.enhance(1.2)  # Adjust the enhancement factor as needed

        # Appending the preprocessed image to the list
        preprocessed_images.append(img)


```

```python
import numpy as np
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.preprocessing import image
from keras.models import Model
from PIL import ImageOps

#  pre-trained VGG16 model
base_model = VGG16(weights='imagenet')

#  new model with the last fully connected layer as the output
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc2').output)

#  list to store extracted features
extracted_features = []

# Loop through preprocessed images
for img in preprocessed_images:
    # Convert the PIL image to a NumPy array
    img_array = image.img_to_array(img)

    # Ensuring the image has three color channels (RGB)
    if img_array.shape[-1] == 1:
        img_array = ImageOps.grayscale(img).convert('RGB')
        img_array = image.img_to_array(img_array)

    # Resizing  the image to match VGG16 input shape (224x224)
    img_array = image.smart_resize(img_array, (224, 224))

    # Expanding the dimensions to match VGG16 input shape (224x224)
    img_array = np.expand_dims(img_array, axis=0)

    # Preprocessing the image for VGG16 model
    img_array = preprocess_input(img_array)

    # Using the VGG16 model to extract features
    features = model.predict(img_array)

    # Appending the extracted features to the list
    extracted_features.append(features)

# Converting the list of extracted features to a NumPy array
feature_matrix = np.vstack(extracted_features)

```

```python
import os
import numpy as np
from PIL import Image
from keras.models import Sequential
from keras.layers import Dense

# Directory path containing test images
test_images_directory = "/content/drive/MyDrive/test_images"

# Initializing an empty list to store the extracted features and filenames
extracted_features = []

# Loop through all files in the directory
for filename in os.listdir(test_images_directory):
    # Check if the file is an image (you can adjust the extensions as needed)
    if filename.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
        # Construct the full path to the image file
        image_path = os.path.join(test_images_directory, filename)

        # Load the image and preprocess it for feature extraction
        img = Image.open(image_path)
        img = img.resize((224, 224))  # Resize to the same dimensions used during training
        img = img.convert('RGB')  # Ensure it's an RGB image
        img_array = np.array(img)

        # Flatten the image data to create a feature vector
        img_feature = img_array.flatten()

        extracted_features.append((filename, img_feature))


# Creating a new Sequential model for predictions
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(150528,)))  # Update input shape
model.add(Dense(5, activation='softmax'))  # Assuming 5 classes for diabetic retinopathy



# Making predictions using the extracted features
predicted_labels = []

for filename, features in extracted_features:
    features = np.array(features)
    features = features.reshape(1, -1)  # Reshape to (1, 150528) to match the updated input shape
    predicted_label = model.predict(features)
    predicted_labels.append((filename, np.argmax(predicted_label)))

# Printing the results
for filename, predicted_label in predicted_labels:
    print(f"Filename: {filename}, Predicted Label: {predicted_label}")
```

```python
import os
import pandas as pd
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.preprocessing import image
import numpy as np

# Define the directory path containing test images
test_images_directory = "/content/drive/MyDrive/test_images"

# Loading the pre-trained VGG16 model
model = VGG16(weights='imagenet', include_top=True)

# Making predictions using the VGG16 model
predicted_labels = []

# Listing image files in the test directory
image_files = [f for f in os.listdir(test_images_directory) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))]

for image_file in image_files:
    # Constructing the full path to the image file
    image_path = os.path.join(test_images_directory, image_file)

    # Loading and preprocess the image
    img = image.load_img(image_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)

    # Predicting
    predictions = model.predict(img)
    decoded_predictions = decode_predictions(predictions, top=1)[0]  # Get the top prediction
    label = decoded_predictions[0][1]  # Get the label of the top prediction

    # Extracting the 'id_code' from the filename
    id_code = image_file.split('_')[0]

    predicted_labels.append((id_code, label))

# Creating a DataFrame with id_code and predicted labels
df = pd.DataFrame(predicted_labels, columns=['id_code', 'predicted_label'])

# Printing the first 10 rows of the DataFrame
print(df.head(10))
```
